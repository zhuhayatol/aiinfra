```python
def get_status(tokens):
    counts = {}
    for i, j in zip(tokens[:-1], tokens[1:]):
        counts[(i,j)] = counts.get((i,j), 0) + 1
    counts = dict(sorted(counts.items(), key=lambda item: item[1], reverse=True))
    return counts

counts = get_status(tokens)
print(counts)
print(len(counts))
a = bytes([229, 138, 160])
print(a.decode("utf-8"))
```
